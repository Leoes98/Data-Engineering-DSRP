## Exercise:
Create a MapReduce algorithm in python to process the william_sh.txt file

```python
from collections import Counter

class MapReduce:
  def __init__(self, file_txt):
    input_txt = open(file_txt, "r")
    content_list = input_txt.read()
    self.input = content_list

  def splitting(self):
    self.splitted = ' '.join([i.lower() for i in self.input.split() if i.isalpha()]).split()
    return self.splitted

  def mapping(self):
    big_map = []
    for split in self.splitted:
      big_map.append((split,1))
    self.mapped = big_map
    return self.mapped

  def shuffling(self):
    self.shuffle = sorted(self.mapped,  key=lambda x: x[0])
    return self.shuffle
  
  def reducing(self):
    self.reduced = dict(Counter(elem[0] for elem in self.shuffle))
    return self.reduced
```

```python
william_sh_txt = MapReduce("william_sh.txt")
william_sh_txt.splitting()
william_sh_txt.mapping()
william_sh_txt.shuffling()
william_sh_txt.reducing()
```
